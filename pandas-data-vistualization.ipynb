```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import seaborn as sns
```

```python
import pandas as pd
import numpy as np
```

## Line Chart

```python
df = pd.DataFrame(np.random.randn(1000), index=pd.date_range('2021-01-01', periods=1000), columns=["Values"])

series = pd.Series(np.random.randn(1000), index=pd.date_range('2022-01-01', periods=1000))
```

```python
df.plot()
```

```python
series.plot()
```

```python
df['Values'] = df['Values'].cumsum()

df.plot()
```

```python
series = series.cumsum()

series.plot(figsize=(20, 5))
```

```python
secondary_axis = df.plot()
series.plot(figsize=(15, 10),  secondary_y=True, ax=secondary_axis)
```

## Bar Chart

```python
df['Values 2'] = series
# df.iloc[999].plot(kind='bar')
df.iloc[999].plot.bar(color='r', alpha=0.6)
```

```python
df.tail().plot(kind="bar", stacked=True)
```

```python
df.tail().plot.barh(stacked=True)
```

# Histogram

```python
iris = sns.load_dataset('iris')
iris.plot.hist(bins=40)
```

```python
iris.plot.hist(bins=40, stacked=True, figsize=(20, 5))
```

```python
iris.plot.hist(bins=40, stacked=True, figsize=(20, 5), orientation="horizontal")
```

```python
iris[['sepal_length', 'sepal_width', 'petal_width', 'petal_length']].diff().hist(bins=45, figsize=(20, 5))
```
